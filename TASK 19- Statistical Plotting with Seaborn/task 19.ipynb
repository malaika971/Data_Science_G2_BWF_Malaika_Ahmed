## ðŸŸ¡ **Linear Regression Plot** 
- Linear Regression plot draw a scatterplot of two variables, x and y.
- Fits the regression model y ~ x
- Plot the resulting regression line and a 95% confidence interval for that regression

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df3=sns.load_dataset("iris")
```

## ðŸŸ¡ **Residual Plot**
- The residplot() function can be a useful tool for checking whether the simple regression model is appropriate for a dataset. 
- It fits and removes a simple linear regression and then plots the residual values for each observation.
- Ideally, these values should be randomly scattered around y = 0

```python
sns.residplot(data=df3,x='sepal_length',y='petal_length')
plt.show()
```

## ðŸŸ¡ **Histogram**
- Used generally for numerical univariate data
- Shows the distribution of a dataset, showing the frequency of data points within specified ranges (bins).
- Its plot is "Axse level" and not "Figure level"


```python
pd.read_csv("world-data-2023.csv")
```

```python
df=pd.read_csv("world-data-2023.csv")
```

```python
df.isnull().sum().sum()
```

```python
df.duplicated().sum()
```

```python
sns.histplot(data=df,x='Birth Rate')
plt.title("HISTOGRAM \n Birth Rate around the world per \n WHO statistics - 2023",fontsize=14,pad=20)
plt.xlabel("Birth Rate",fontsize=14,labelpad=25)
plt.ylabel("Frequency",fontsize=14,labelpad=25)
plt.show()
```

## ðŸŸ¡ **Kernel Density Estimate (KDE) Plot**
-  Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimat.
 - 
  Works just like histogrut instead of creating bins,it creates a cu.rve




```python
# DENSITY on y-axis is PROBABILITY

sns.kdeplot(data=df,x='Birth Rate')
plt.title("KDE  \n Birth Rate around the world per \n WHO statistics - 2023",fontsize=14,pad=20)
plt.xlabel("Birth Rate",fontsize=16,labelpad=15)
plt.ylabel("Density",fontsize=16,labelpad=15)
plt.show()
```

## ðŸŸ¡ **Distribution Plot**

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency

```python
# dist is figure level function
# shape is square by default
```

```python

sns.displot(data=df,x='Birth Rate',kind='hist')
plt.title("Distribution plot: Histogram \n Birth Rate around the world  \n  WHO statistics - 2023",fontsize=14,pad=20)
plt.xlabel("Birth Rate",fontsize=14,labelpad=25)
plt.ylabel("Frequency",fontsize=14,labelpad=25)

plt.show()
```

```python
# Changing the kind of plot in distribution plot
```

```python
sns.displot(data=df,x='Birth Rate',kind='kde')
plt.title("Distribution plot: KDE \n Birth Rate around the world \n  WHO statistics - 2023",fontsize=14,pad=25)
plt.xlabel("Birth Rate",fontsize=16,labelpad=15)
plt.ylabel("Density",fontsize=16,labelpad=15)
plt.show()
```

```python
## Filling the curve
```

```python
sns.displot(data=df,x='Birth Rate',kind='kde',fill=True)
plt.title("Distribution plot: KDE \n Birth Rate around the world \n  WHO statistics - 2023",fontsize=14,pad=20)
plt.xlabel("Birth Rate",fontsize=16,labelpad=15)
plt.ylabel("Density",fontsize=16,labelpad=15)
plt.show()
```

```python
# FIGURE LEVEL PLOT

dfn=df1[df1['Country'].isin(['Pakistan','Austria','India','Ireland'])]
plt.figure(figsize=(8,5))

sns.catplot(data=dfn, x='Country', y='Life_expectancy',kind='violin',hue='Region')

plt.title('Distribution Plot: Violin \n Average life expectency \n WHO Statistics 2000-2015)',fontsize=14,pad=20)

plt.xlabel('Country', labelpad=15,fontsize=14)
plt.ylabel('Life Expectency', labelpad=15,fontsize=14)

plt.show()
```

## ðŸŸ¡ **Rug Plot**

- Plot marginal distributions by drawing ticks along the x and y axes.

- This function is intended to complement other plots by showing the location of individual observations

```python
sns.kdeplot(data=df,x='Birth Rate')
sns.rugplot(data=df,x='Birth Rate')

plt.title("KDE+Rug Plot  \n Birth Rate around the world \n  WHO statistics - 2023",fontsize=14,pad=20)
plt.xlabel("Birth Rate",fontsize=16,labelpad=15)
plt.ylabel("Density",fontsize=16,labelpad=15)

plt.show()
```

## ðŸŸ¡ **Box Plot**
- Standardized way of displaying the distribution of data based on a five number summary (â€œminimumâ€, first quartile [Q1], median, third quartile [Q3] and â€œmaximumâ€).
- It can tell you about your outliers and what their values are.
- Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed

```python
dfd=df1[df1['Country'].isin(['Pakistan','Austria','India','Ireland'])]

sns.boxplot(data=dfd,x='Country',y='Life_expectancy')
```

```python

sns.boxplot(data=dfd,x='Country',y='Life_expectancy',hue='Region')
plt.title('Comparison of Average Life Expectency \n in 4 countries \n (Acc to Statistics 2000-2015)',fontsize=12,pad=20)

plt.xlabel('Country', labelpad=15,fontsize=14)
plt.ylabel('Life Expectency', labelpad=15,fontsize=14)
```

## ðŸŸ¡ **JOINT PLOT**
- Joint plot shows the multi-panel figure that shows both the bivariate( or joint) relatinship between two variables along with the univariate ( or marginal ) distribution of each on separate axis.

```python
g=sns.jointplot(data=df2,x='total_bill',y='tip',height=5)
g.fig.suptitle("JOINT PLOT: Scatter and Histogram",fontsize=15)
g.fig.subplots_adjust(top=0.9)
```

```python
g1=sns.jointplot(data=df2,x='total_bill',y='tip',kind='kde',height=5)
g1.fig.suptitle("JOINT PLOT: 1D and 2D KDE",fontsize=15)

plt.figure(figsize=(2,2))
g1.fig.subplots_adjust(top=0.9)
```

## ðŸŸ¡ **FACET GRID**
- FacetGrid allows for plotting subsets of data across multiple plots, making it easy to compare different groups.
- It helps in exploring relationships between multiple variables by plotting them across different categories.
- FacetGrid provides a clean, efficient layout to visualize complex data without overcrowding a single plot.

```python

fg=sns.FacetGrid(df2)
fg.map(plt.scatter,'total_bill','tip')

fg.fig.suptitle('Distribution of Total Bill', fontsize=14)

fg.fig.subplots_adjust(top=0.8)
plt.show()
```

```python
# FACET GRID SHOWING THE TWO PLOTS DISTRIBUTED ON GENDER

fg=sns.FacetGrid(df2,col='Gender')
fg.map(plt.scatter,'total_bill','tip')
fg.fig.suptitle('Distribution of Total Bill', fontsize=18)
fg.fig.subplots_adjust(top=0.7)


plt.show()
```

```python
# FACET GRID SHOWING PLOTS DISTRIBUTED ON GENDER AND DAYS

fg=sns.FacetGrid(df2,col='Gender',hue='day',hue_order=['Thur','Fri','Sat','Sun'])
fg.map(plt.scatter,'total_bill','tip').add_legend()
fg.fig.suptitle('Distribution of Total Bill', fontsize=18)
fg.fig.subplots_adjust(top=0.7)

plt.show()
```

```python
# FACET GRID SHOWING PLOTS DISTRIBUTED ON DAYS

fg=sns.FacetGrid(df2,col='day',hue='Gender')
fg.map(plt.scatter,'total_bill','tip').add_legend()

fg.fig.suptitle('Distribution of Total Bill', fontsize=20)
fg.fig.subplots_adjust(top=0.6)

plt.show()
```

```python
# FACET GRID (BAR PLOT) SHOWING PLOTS DISTRIBUTED ON DAYS 

fg=sns.FacetGrid(df2,col='day',hue='Gender')
fg.map(plt.bar,'total_bill','tip').add_legend()

fg.fig.suptitle('Distribution of Total Bill', fontsize=20)
fg.fig.subplots_adjust(top=0.6)


plt.show()
```
