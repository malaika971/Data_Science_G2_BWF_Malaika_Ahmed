```python
df=pd.read_csv('weatherdata.csv')
```

#### ðŸ”¹ Extracting mini data set

```python
plt.plot(x,y,)
plt.title('SIMPLE LINE PLOT')
plt.xlabel('Wind_Speed(kmh)')
plt.ylabel('TEMPERATURE (C)')

plt.show()
```

### 2. Plot multiple lines on the same graph


```python
x=df1['Wind_Speed_kmh']
y1=df1['Temperature_C']
y2=df1['Humidity_pct']
y3=df1['Precipitation_mm']

plt.plot(x,y1,label='Temp & humidity')
plt.plot(x,y2,label='Temp and windspeed')
plt.plot(x,y3,label='Temp and precipitation')

plt.title('LINE PLOT with multiple lines')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.legend( )
plt.show()
```

### 3. Add markers to a line plot

#### ðŸ”¹ LINE plot with marker style :circle

```python
x=df1['Wind_Speed_kmh']
y=df1['Temperature_C']

plt.plot(x,y,marker='o',ms='10')
plt.title('LINE plot with marker style :circle')
plt.xlabel('Wind_Speed(kmh)')
plt.ylabel('TEMPERATURE (C)')

plt.show()
```

 #### ðŸ”¹ LINE PLOT with marker style : Diamond

#### ðŸ”¹ LINE PLOT with marker style :Triangle

### Custom Marker edge colors

```python
plt.plot(x,y,marker='o',ms='10',mec='r')
plt.title('LINE PLOT \n Marker style: Circle \n Marker edge color: Red')
plt.xlabel('Wind_Speed(kmh)')
plt.ylabel('TEMPERATURE (C)')

plt.show()
```

```python

# Different color and different size of each point with colour bar
df1=df[0:10]

x=df1['Wind_Speed_kmh']
y=df1['Temperature_C']
color=['red','blue','green','brown','orange','black','yellow','red','green','black']
size=[180,150,170,125,140,250,160,280,155,200]
plt.scatter(x,y,marker='.',c=color,s=size)
plt.title('Scatter Plot \n Marker Style: Point \n Different sizes and colors for each point \n With color bar')
plt.xlabel('Wind_Speed(kmh)')
plt.ylabel('TEMPERATURE (C)')
plt.colorbar()
plt.show()
```

## ðŸŸ¡ Bar Charts 


### `1. Create a vertical bar chart.


```python
df1=df[20:25]
df2=df[25:30]

x=df1['Location']
y=df2['Temperature_C']  

plt.bar(x,y)
plt.title('Vertical Bar Chart')
plt.xlabel('Location')
plt.ylabel('TEMPERATURE (C)')

plt.show()





```

### 2. Create a horizontal bar chart


```python


x=df1['Temperature_C']
y=df2['Location']  

plt.barh(x,y)
plt.title('Horizontal Bar Chart')
plt.xlabel('Location')
plt.ylabel('TEMPERATURE (C)')

plt.show()
```

### 3 Customize bar colors and add edge colors.


```python
x=df1['Location']
y=df2['Temperature_C']  
c=['blue','green','red','black','orange']
w=[0.2,0.3,0.5,0.4,0.6]
plt.bar(x,y,color=c,width=w)
plt.title('Vertical Bar Chart \n Different colored bars with varying width')
plt.xlabel('Location')
plt.ylabel('TEMPERATURE (C)')

plt.show()
```

```python

x=df1['Location']
y=df2['Temperature_C']  
c=['blue','green','red','black','orange']
h=[0.2,0.3,0.5,0.4,0.6]
plt.barh(x,y,color=c,height=h)
plt.title('Horizontal Bar Chart \n Different colored bars with varying height')
plt.xlabel('Temperature')
plt.ylabel('Location')

plt.show()


```

### Adding edge colors

```python
x=df1['Location']
y=df2['Temperature_C']  

plt.bar(x,y,color='yellow',edgecolor='red',width=0.2)
plt.title('Vertical Bar Chart \n Edge color: Red')
plt.xlabel('Location')
plt.ylabel('TEMPERATURE (C)')

plt.show()
```

```python

x=df1['Temperature_C']
y=df2['Location']  

plt.barh(x,y,color='yellow',edgecolor='black',height=0.5)
plt.title('Horizontal Bar Chart \n Edge Color : Black')
plt.xlabel('Location')
plt.ylabel('TEMPERATURE (C)')

plt.show()
```

## ðŸŸ¡ Pie Charts


### 1. Create a basic pie chart


```python
df1=df[20:25]
df2=df[25:30]

# Create a vertical bar chart.
x=df1['Location']
y=df2['Wind_Speed_kmh']  

plt.pie(y)
plt.title('Pie Chart')


plt.show()
```

## 2.Customize slice colors and add a legend.


```python
df1=df[20:25]
df2=df[25:30]

x=df1['Location']
y=df2['Wind_Speed_kmh']  

plt.pie(y,labels=x)
plt.title('Pie Chart with labels')

plt.legend()
plt.show()
```

```python
df1=df[20:25]
df2=df[25:30]

x=df1['Location']
y=df2['Wind_Speed_kmh']  

Color=['red','blue','green','yellow','orange']
plt.pie(y,labels=x,startangle=90,colors=Color)
plt.title('Pie Chart with customized colors and labels')

plt.legend(title='LOCATION')
plt.show()
```

## 3.Explode a slice of the pie chart.


```python
df1=df[20:25]
df2=df[25:30]

x=df1['Location']
y=df2['Wind_Speed_kmh']  

Color=['red','blue','green','yellow','orange']
plt.pie(y,labels=x,startangle=90,colors=Color,explode=[0,0.5,0,0,0])
plt.title('Pie Chart with explode blue slice ')

plt.show()
```

```python
df1=df[20:25]
df2=df[25:30]

x=df1['Location']
y=df2['Wind_Speed_kmh']  

Color=['red','blue','green','yellow','orange']
plt.pie(y,labels=x,startangle=90,colors=Color,explode=[0.2,0.5,0,0.2,0])
plt.title('Pie Chart with explode blue,red and yellow slice ')

plt.show()
```

## 4. Add percentage labels to slices


```python
df1=df[20:25]
df2=df[25:30]

x=df1['Location']
y=df2['Wind_Speed_kmh']  

Color=['red','blue','green','yellow','orange']
plt.pie(y,labels=x,startangle=90,colors=Color,autopct='%2.1f%%')
plt.title('Pie Chart with % labels')

plt.show()
```
