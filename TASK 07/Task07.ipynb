```python
### _______________________________________________________________________________
# ==================================STEP #01====================================
# ================================CLASS "Book"==================================
#_______________________________________________________________________________

# Defining the class Book,which stores different objects(books in our case)
# ,its parameters(title,author,isbn and available status) and 
# ,methods i.e display_info,mark_as_borrowed,mark_as_returned

class Book:

    # Constructor function __init__,which carries parameters title,author and isbn.
    def __init__(self, title, author, isbn):
     
       # self.title is attribute while title is parameter
        self.title = title                 
        self.author = author               
        self.isbn = isbn
        self.status = 'available'


    
    # Defining method to display the information(attritubes) of particular book
    def display_info(self):

      # Inside function storing the print function to display 
      # the value of parameters (stored in attributes while defining class "Book") 
      # using string contatenation method.When we will call display_info,it will displays
      # all the info about that specific book,its title,author
      # isbn and whether its available or borrowed'''
        print("Title: " + self.title 
              + ", Author: " + self.author 
              + ", ISBN: " + self.isbn 
              + ", Status: " + self.status)



    # Defining method to mark the status of particular book as borrowed
    def mark_as_borrowed(self):
        # Setting the status of book as borrowed
        self.status = 'borrowed'
        


    # Defining method to mark the status of particular book as available
    def mark_as_returned(self):
        #Setting the status of book as availble once its returned
        self.status = 'available'


#_______________________________________________________________________________
# ===============================STEP #02=======================================
# ============================CLASS "Member"====================================
#_______________________________________________________________________________

# Defining the class Member,which stores different objects(data of members)
# ,its parameters(member_name and member_id) 
# ,and methods borrow_book,return_book,display_info
class Member:
      # Constructor_function with parameters,name and ID
      def __init__(self,name,member_id):
           self.name=name
           self.member_id=member_id
           # initialized an empty list of attribute self.borrowed_books
           # ,which allows to store the data of books borrowed by member
           self.borrowed_books=[]
      


      # Method borrow_book allows the member to borrow book.
      # It checks if book is available
      # Then add this book in list of borrowed_books
      # Calls the mark_as_borrowed method to update its status
      def borrow_book(self, book):
         if book.status == 'available':
            self.borrowed_books.append(book)
            book.mark_as_borrowed()
            print("Book borrowed successfully.")
         else:
            print("Book is already borrowed.")



      # Method return_book allows the member to return book.
      # It checks if book is in borrowed_books of that member
      # Then removes this book from the list of borrowed_books
      # Calls the mark_as_returned method to update its status
      def return_book(self, book):
        if book in self.borrowed_books:
           self.borrowed_books.remove(book)
           book.mark_as_returned()
           print("Book returned successfully.")
        else:
           print("Book is not borrowed by this member.")



      # Method display_info displays the data of member
      # Its name,ID and history of borrowed books
      def display_info(self):
          print("Name:", self.name)
          print("Member ID:", self.member_id)
          print("Borrowed Books:")
          if self.borrowed_books: 
            # for loop will iterate through each book that is borrowed by member
            # ,stored in borrowed_books and prints the title of that books   
            for book in self.borrowed_books:
                print(f"- {book.title}")
          else:
              print("No books borrowed.")  
     
#_______________________________________________________________________________
# ===============================STEP #03=======================================
# ============================CLASS "Library"===================================
#_______________________________________________________________________________

# Defining the class Library,which stores different objects(data of books and members) 
# and methods register_member,issue_book,return_book,display_all_books,display_all_members

class Library:
    def __init__(self):
        self.books = []
        self.members = []




    # Defining method add_book to add new book to the library
    def add_book(self,title,author,isbn):
        # Creating an object "book" within class "BooK"
        book = Book(title, author, isbn)
        # Appending tha object "book" into list "books"
        self.books.append(book)
        print("Book " + title + " added to the library.")




    # Defining method "register_member" to register new member to the library
    def register_member(self,name,member_id):
        # Creating an object "member" within class "Member"
        member = Member(name, member_id)
        # Appending tha object "member" into list "members"
        self.members.append(member)
        print("Member " + name + " registered with ID " + member_id + ".")

 



    # Defining method "issue_book" to issue book to the member
    def issue_book(self, member_id, isbn):
        print(f"Issuing book with ISBN {isbn} to member with ID {member_id}")  

    # Finding the member with the given member_id
    # setting member as None (starting with assumption no member found yet)
        member = None

        # for loop will Iterate through each member in the list "members"
        for m in self.members:
            # Checking if the given member_id matches with member_id already present in list
              if m.member_id == member_id:
                 member = m
                 break

    # Finding the book with the given ISBN
    # setting book as None (starting with assumption no book found yet)
        book = None
    # for loop will Iterate through each book in the list "books"
        for b in self.books:
    # Checking if the given isbn matches with isbn already present in list
            if b.isbn == isbn:
               book = b
               break  

    # Issue the book to the member if both member and book are found
            if member is not None and book is not None:
                   print(f"Issuing book with ISBN {isbn} to member with ID {member_id}")
    # In class "Member" with object "member",method "borrow_book" will allow member
    # to borrow book by passing argument "book"(book is object)
                   member.borrow_book(book)
                   book.mark_as_borrowed()
                   book.display_info()
            else:
                 print("Cannot return book. Member ID"+ member_id + "and ISBN" + isbn + "not found.")



    # Defining method return_book to add book into library, returned by member
    def return_book(self, member_id, isbn):
    # Find the member with the given member_id
        member = None
        for m in self.members:
            if m.member_id == member_id:
               member = m
               break
    # Find the book with the given ISBN
        book = None
        for b in self.books:
            if b.isbn == isbn:
               book = b
               break
    # Return the book if both member and book are found
        if member and book:
            member.return_book(book)
        else:
            print("Cannot return book. Member ID " + member_id + " and ISBN " + isbn + " not found.") 



    # Defining method issue_book to issue book to the member 
    def display_all_books(self):
        print("All Books in Library:")
        # for loop will iterate through each book stored in list "books"
        # and it will print info of that book by calling method display_info
        # defined in class "Book" 
        for book in self.books:
            book.display_info()


    
    
    # Defining method issue_book to issue book to the member  
    def display_all_members(self):
        # for loop will iterate through each member stored in list "members"
        # and it will print info of that member by calling method display_info
        # defined in class "Member"
        print("All Members of Library:")
        for member in self.members:
            member.display_info()   



#_______________________________________________________________________________
# ================================STEP #04======================================
# ========================USER_INPUT_FUNCTIONS==================================
#_______________________________________________________________________________
 
# Function to add a book based on user input
def add_book_from_input():
    title = input("Enter book title: ")
    author = input("Enter book author: ")
    isbn = input("Enter book ISBN: ")
    return Book(title, author, isbn)

# Function to register a member based on user input
def register_member_from_input():
    name = input("Enter member name: ")
    member_id = input("Enter member ID: ")
    return Member(name, member_id)

# Function to issue a book based on user input
def issue_book_from_input(library):
     member_id = input("Enter member ID: ")
     isbn = input("Enter book ISBN: ")
     library.issue_book(member_id, isbn)

# Function to return a book based on user input
def return_book_from_input(library):
     member_id = input("Enter member ID: ")
     isbn = input("Enter book ISBN: ")
     library.return_book(member_id, isbn)

#_______________________________________________________________________________
# ===============================STEP #05=======================================
# ===========================MAIN_FUNCTION======================================
#_______________________________________________________________________________

# Main program for creating instances adding books, registering members, 
# and simulating borrowing/returning books.

def main():
    # Creating an instance of the library
    library = Library()

    # Adding some initial books
    initial_books = [
        Book("Club of 5am", "Robin Sharma", "ISBN001"),
        Book("And the mountains echoed", "Khalid Hosani", "ISBN002"),
        Book("Revive your heart", "Nouman Ali khan", "ISBN003")
    ]                        
    # for loop will iterate through each book stored in "initial_books"
    # and it will call method "add_book" to instance "library"
    for book in initial_books:
        library.add_book(book.title, book.author, book.isbn) 


    # Registering some initial members
    initial_members = [
        Member("Zainab", "M001"),
        Member("Fatimah", "M002")
    ]
    # for loop will iterate through each member stored in "initial_members"
    # and it will call method "register_member" to instance "library"
    for member in initial_members:
        library.register_member(member.name, member.member_id)
         


 # Menu to interact with the library management system
    while True:
        print("\n______Library Management System Menu_______\n")
        print("1. Add a Book")
        print("2. Register a Member")
        print("3. Issue a Book to a Member")
        print("4. Return a Book from a Member")
        print("5. Display All Books")
        print("6. Display All Members")
        print("7. Exit")

        choice = input("Enter your choice (1-7): ")

        if choice == '1':
           book = add_book_from_input()
           library.add_book(book.title,book.author,book.isbn)

        elif choice == '2':
             member = register_member_from_input()
             library.register_member(member.name,member.member_id)

        elif choice == '3':
             issue_book_from_input(library)

        elif choice == '4':
             return_book_from_input(library)

        elif choice == '5':
             library.display_all_books()

        elif choice == '6':
             library.display_all_members()

        elif choice == '7':
             print("Exiting the program. Goodbye!")
             break

        else:
             print("Invalid choice. Please enter a number from 1 to 7.")

if __name__ == "__main__":
    main()    
 
    
   

```
