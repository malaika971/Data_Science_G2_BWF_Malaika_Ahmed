 ## <span style='color:Blue'>  ***2. Removing Punctuation & Special Characters*** </span>

- Punctuation removal is a text preprocessing step where you remove all punctuation marks (such as periods, commas, exclamation marks, emojis etc.) from the text to simplify it and focus on the words themselves.
- In Handling special characters and punctuation,typically we may choose to remove them or replace them with a space, depending on your specific requirements.

```python
import string,time
string.punctuation
```

#### These are the characters which python consider as punctuations

```python
exclude = string.punctuation
```

### ***Creating a function to remove punctuations from text***

```python
# str.maketrans('', '', exclude) creates a plan or a "map" that tells Python which characters to remove.
# The first two '' (empty strings) mean weâ€™re not changing any characters to something else.
# The exclude part is telling Python, "Remove these characters.
# text.translate(...) uses the map to remove the unwanted characters from the text.


def remove_punc1(text):
    return text.translate(str.maketrans('', '', exclude))
```

 ## <span style='color:Blue'>  ***3. Removal of HTML Tags*** </span>

Removal of HTML Tags is a text preprocessing step used to clean text data from HTML documents. When working with text data obtained from web pages or other HTML-formatted sources, the text may contain HTML tags, which are not desirable for text analysis or machine learning models. Therefore, itâ€™s important to remove HTML tags from the text data.

***
### ***Importing re***

***re is "regular expressions," which is a powerful tool in Python for matching patterns in text.This module helps to find and manipulate patterns, such as removing specific types of characters or strings from a text.***

```python
import re
```

### ***Creating function to remove html tags***

```python
# re.compile('<.*?>') creates a regular expression pattern to match anything that looks like an HTML tag.
# pattern.sub(r'', text) uses the pattern created in the previous step to search for all occurrences of HTML tags in text.
# sub() is a method that substitutes (replaces) parts of the text that match the pattern.
# r'' means "replace with an empty string," effectively removing the HTML tags.
# So, any text that matches the pattern (i.e., any HTML tag) will be replaced by nothing, effectively removing it from the text.

def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)
```

### ***Applying the function on review column of dataset***

```python
df['review'] = df['review'].apply(remove_html_tags)
```

 ## <span style='color:Blue'>  ***4. Removal of URLs*** </span>

When building a model, URLs are typically not relevant and can be removed from the text data.


### ***Creating function to remove html tags***

```python
# re.compile(): This function compiles a regular expression pattern into a regex object, which can be used to match strings.
# r'https?://\S+|www\.\S+': This is the regular expression pattern used to match URLs
# sub() is a method that replaces all occurrences of the pattern found in text with a specified replacement.
# r'': This is an empty string. By replacing the matched URLs with an empty string, they are effectively removed from the text.

def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

 ## <span style='color:Blue'>  ***6. Removing Stop Words*** </span>
- Stopwords refer to the most commonly occurring words in any natural language.
- For the purpose of analyzing text data and building NLP models, these stopwords might not add much value to the meaning of the document. Therefore, removing stopwords can help us to focus on the most important information in the text and improve the accuracy of our analysis.
- One of the advantages of removing stopwords is that it can reduce the size of the dataset, which in turn reduces the training time required for natural language processing models.

image.png

***
- import nltk: Imports the Natural Language Toolkit (NLTK) library, which provides tools for working with human language data.
- nltk.download('stopwords'): Downloads the stopwords dataset from NLTK. Stopwords are common words (e.g., "and", "the", "in") that are often removed from text during preprocessing.

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
```

- new_text = []: Initializes an empty list new_text to store the words that are not stopwords.

- for word in text.split(): Splits the input text into individual words and iterates over each word.

- if word in stopwords.words('english'):: Checks if the current word is in the list of English stopwords.

- stopwords.words('english'): Retrieves the list of English stopwords from NLTK.

- new_text.append(''): If the word is a stopword, append an empty string to new_text. This effectively removes the stopword from the resulting text.

- else: new_text.append(word): If the word is not a stopword, append it to new_text.

- x = new_text[:]: Creates a copy of new_text and assigns it to x. This is done to preserve the original list.

- new_text.clear(): Clears the new_text list. This is unnecessary since the list new_text is not used after this point.

- return " ".join(x): Joins the words in x into a single string with spaces in between, effectively reconstructing the text without stopwords.

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

## <span style='color:Blue'>  ***9. Stemming and Lemmatization*** </span>


Stemming involves reducing words to their root form, while lemmatization reduces them to their base or dictionary form. Both processes aim to unify different word forms to streamline text analysis by treating variations of words as a single entity, facilitating more accurate and efficient language processing.




image.png

## ***Stemming***



```python
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Initialize the PorterStemmer
stemmer = PorterStemmer()

# Example sentences
sentences = [
    'I am pursuing MS in A.I. from NUST',
    "You may email me at ma@gmail.com",
    'I travel 5km to university'
]
```

```python
# Function to perform stemming
def stem_sentences(sentences):
    for sentence in sentences:
        print(f"\nOriginal Sentence: {sentence}")

        # Tokenize the sentence into words
        tokens = word_tokenize(sentence)

        # Apply stemming to each word
        stemmed_words = [stemmer.stem(word) for word in tokens]

        # Join the stemmed words back into a sentence
        stemmed_sentence = ' '.join(stemmed_words)
        print(f"Stemmed Sentence: {stemmed_sentence}")

# Call the function
stem_sentences(sentences)
```

## ***Lemmatization***

```python
import spacy

# Load the spaCy model for lemmatization
nlp = spacy.load("en_core_web_sm")
```

```python
# Example sentences
sentences = [
    'I am pursuing MS in A.I. from NUST',
    "You may email me at ma@gmail.com",
    'I travel 5km to university'
]
```

```python
# Function to perform lemmatization
def lemmatize_sentences(sentences):
    for sentence in sentences:
        print(f"\nOriginal Sentence: {sentence}")

        # Process the sentence using spaCy's NLP model
        doc = nlp(sentence)

        # Extract the lemma (base form) for each word in the sentence
        lemmatized_words = [token.lemma_ for token in doc]
        print(f"Lemmatized Words: {lemmatized_words}")

# Call the function
lemmatize_sentences(sentences)
```
