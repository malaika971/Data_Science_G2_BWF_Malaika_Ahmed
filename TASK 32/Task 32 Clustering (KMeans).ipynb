 # <span style='color:Red'>  ***What is Clustering ?*** </span>



Clustering is grouping data based on similarity patterns based on distance




## ***What is K-means ?***
- K-means clustering is a simple and elegant approach for partitioning a
data set into K distinct, non-overlapping clusters. To perform K-means
clustering, we must first specify the desired number of clusters K; then the
K-means algorithm will assign each observation to exactly one of the K
clusters 

- The K-means algorithm clusters data by trying to separate samples in n groups of equal variance, minimizing a criterion known as the inertia or within-cluster sum-of-squares. The K-means algorithm aims to choose centroid that minimise the inertia, or within-cluster sum-of-squares criterion 


## ***K Mean Clustering***
- K-means clustering is an unsupervised machine learning algorithm, meaning it learns from input data without labeled examples or explicit guidance.

- It divides a dataset into “K” distinct, non-overlapping groups called clusters. Each cluster contains data points with similar characteristics, and each cluster is different from the others.

- The grouping is done by minimizing the sum of squares of distances between data and the corresponding cluster centroid. The main advantage of K-Means is its simplicity. Its algorithm is easy to understand, and it’s highly flexible to any changes or anomalies in the given data.
- Its simplicity and efficiency make it a popular choice for various applications, such as customer segmentation, image compression, anomaly detection, and social network analysis.

 # <span style='color:Red'>  ***How does the K-Means Algorithm Work?*** </span>


The K-Means algorithm is an iterative process that employs a series of simple, sequential steps to achieve its primary goal of clustering data. Let’s understand the key stages involved in its implementation:

### ***Step 1: Select the number of clusters (K)***
This is the ‘K’ in K-Means. K represents the number of clusters you want to create. You can choose K based on your understanding of the data or by using methods like the elbow method, which involves finding the optimal number of clusters that reduces the within-cluster variation. We will understand the elbow method later.

To illustrate the K-Means clustering process, let’s consider a graph with various data points, where each point represents an individual observation or instance within our dataset. For this example, we will assume that we want to create 3 clusters from these data points.



### ***Step 2: Randomly select K distinct data points***
Once you have determined the value of K, the K-Means algorithm randomly selects K distinct data points from the dataset to serve as the initial cluster centroids. A centroid is an imaginary point that represents the center of a cluster. The algorithm will iteratively update the positions of these centroids to minimize the overall distance between the data points and their assigned centroids.




### ***Step 3: Measure the distance between each data point and the K initial centroids***
For each data point in the dataset, the algorithm calculates the distance between the data point and each of the K centroids. The distance metric commonly used is Euclidean distance (d), which measures the straight-line distance between two points.



### ***Step 4: Assign each data point to the nearest cluster***
After calculating the distances, the algorithm assigns each data point to the cluster with the nearest centroid. This means that each data point is associated with the cluster whose centroid is closest to the data point.



### ***Step 5: Calculate the new centroids or mean of each cluster***

Once all data points have been assigned to a cluster, the algorithm calculates the new centroids for each cluster. The new centroid is the mean (or average) of all the data points within that cluster






### ***Step 6: Repeat steps 3–5 until convergence***

The algorithm then repeats the process of assigning data points to the nearest clusters and recalculating the centroids. This iterative process continues until the cluster assignments no longer change significantly between iterations or a specified number of iterations is reached. This is known as the convergence of the algorithm.





## ***Advantages***
- K-Means is easy to understand and implement, making it a popular choice for clustering tasks.
- The algorithm can handle large datasets efficiently and scales well for a relatively small number of clusters.
- K-Means is unsupervised, meaning it can discover patterns in unlabeled data and is applicable across various domains.

## ***Disadvantages***
- The algorithm’s performance can be affected by the initial selection of cluster centroids, which can lead to sub-optimal clustering results.
- K-Means can struggle with noisy data and outliers, as they may distort the cluster centroids and boundaries.
- K-Means assumes that clusters have a convex, spherical shape, which might not be suitable for datasets with non-convex or arbitrarily shaped clusters.

 # <span style='color:Red'>  ***How to Choose the Right Number of Clusters ?*** </span>

Choosing the optimal number of clusters (K) is essential for achieving the best results with K-Means clustering. However, it can be difficult to determine the appropriate number of clusters just by visually examining the dataset so techniques like the Elbow method can assist in finding the optimal number of clusters.


## ***Elbow Method:***
In the Elbow method, you calculate the Within Cluster Sum of Squares (WCSS) error for various values of ‘K’ and plot these values on a line graph.

WCSS measures the compactness of the clusters. It quantifies the variance or spread of the data points within each cluster. A lower WCSS score indicates that the data points are closer to their centroids, implying better clustering. It measures the sum of squared distances between each data point and its nearest cluster centroid, representing the reduction in variation within clusters.

Look for the “elbow” point — the location on the graph where the curve starts to level off or flatten. This elbow point signifies that adding more clusters beyond this value only results in a minimal decrease in WCSS error. Therefore, the K value at the elbow can be considered the optimal number of clusters for your K-Means clustering m


odel.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("student_clustering.csv")
```

## ***Pre-Processing***

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
print("The shape of data is",df.shape)
```

```python
plt.scatter(df['cgpa'],df['iq'])
plt.title("Shape of original data",pad=20,fontsize=20)
```

## ***Putting this data in elbow method to check whether it classifies the same clusters too***

```python
## ELBOW METHOD TO FIND RIGHT NUMBER OF CLUSTERS

wcss = []  # Creating a list of wcss- This list will be used to store the WCSS values computed for each number of clusters.

for i in range(1,11):
    km = KMeans(n_clusters=i)  # Creating an object of KMeans 
    km.fit_predict(df)
    wcss.append(km.inertia_)   # returns the WCSS value for the current number of clusters amd add the WCSS value to the wcss list.
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

## From above graph we can see that curve tends to be more steep as it reaches 4, Therefore, elbow Curves also concludes that number of clusters should be 4.

```python

# Extracting numpy array for X value  
X = df.iloc[:,:].values
km = KMeans(n_clusters=4)

# It will train the data plus will return the value of cluster too
y_means = km.fit_predict(X)
```

```python
y_means
```

```python
# This shows all rows that comes in 3rd cluster and its 1st vlaue

X[y_means == 3,1]
```

```python
# Extracting x and y coordinate of all points that comes under cluster 0
plt.scatter(X[y_means == 0,0],X[y_means == 0,1],color='blue',label="Students with more CGPA and less IQ")

# Extracting x and y coordinate of all points that comes under cluster 1
plt.scatter(X[y_means == 1,0],X[y_means == 1,1],color='red',label="Students with less CGPA and less IQ")

# Extracting x and y coordinate of all points that comes under cluster 2
plt.scatter(X[y_means == 2,0],X[y_means == 2,1],color='green',label="Students with more CGPA and IQ")

# Extracting x and y coordinate of all points that comes under cluster 3
plt.scatter(X[y_means == 3,0],X[y_means == 3,1],color='yellow',label="Students with less CGPA and more IQ")


plt.xlabel('IQ',labelpad=20,fontsize=18)
plt.ylabel('CGPA',labelpad=20,fontsize=18)
plt.title('Scatter Plot with Clusters', pad=20,fontsize=18)
plt.legend()
plt.show()
```

# ***K-Means on 3-D Data***


- ***make_blobs*** is a utility function for generating synthetic datasets with a specific number of clusters. It is often used to create test datasets for clustering algorithms.
- ***centroids*** is a list of tuples where each tuple represents the coordinates of a centroid in a 3-dimensional space. In this case, there are 4 centroids.
- ***cluster_std*** is a list where each value corresponds to the standard deviation of the clusters. In this example, each cluster has a standard deviation of 1, meaning that the points in each cluster will be distributed around the centroid with a spread determined by this standard deviation. A larger standard deviation would result in a more dispersed cluster.
- ***n_features=3:*** Indicates the number of features (dimensions) of the dataset.
- ***random_state=1:*** Sets the seed for the random number generator used by make_blobs. This ensures that the same random data is generated each time the code is run,

## Generating Synthetic Data Set

```python
from sklearn.datasets import make_blobs

centroids = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
cluster_std = [1,1,1,1]

X,y = make_blobs(n_samples=200,cluster_std=cluster_std,centers=centroids,n_features=3,random_state=1)
```

```python
import plotly.express as px
fig = px.scatter_3d(x=X[:,0], y=X[:,1], z=X[:,2])
fig.show()
```

```python
wcss = []
for i in range(1,21):
    km = KMeans(n_clusters=i)
    km.fit_predict(X)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,21),wcss)
```

## From above graph we can see that in curve tends to be more steep as it reaches 4, Therefore, elbow Curves also concludes that number of clusters should be four too in 3D.

```python
km = KMeans(n_clusters=4)
y_pred = km.fit_predict(X)
```

```python
df = pd.DataFrame()

df['col1'] = X[:,0]
df['col2'] = X[:,1]
df['col3'] = X[:,2]
df['label'] = y_pred
```

```python
fig = px.scatter_3d(df,x='col1', y='col2', z='col3',color='label')
fig.show()
```
