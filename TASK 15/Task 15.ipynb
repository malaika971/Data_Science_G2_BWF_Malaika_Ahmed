## 1. Merge two DataFrames on a single key


```python
# Creating first DataFrame

df1=pd.DataFrame({'KEY': ['a','b','c','d','e','f'] ,'Data1':range(6)})
```

```python
df1
```

```python
# creating second DataFrame

df2=pd.DataFrame({'KEY': ['c','d','e'] ,'Data2':[6,7,8]})
```

```python
df2
```

```python
df_merge_s=pd.merge(df1,df2,on='KEY')
df_merge_s
```

## 3. Perform an outer join, inner join, left join, and right join.


```python
# First DataFrame
df5=pd.DataFrame({'lKEY': ['a','b','c','d','e','f'] ,'Data1':range(6)})
```

```python
df5
```

```python
# Second DataFrame
df6=pd.DataFrame({'rKEY': ['c','d','e'] ,'Data2':[6,7,8]})
```

```python
df6
```

```python
# Merging using left,right and inner join
# inner join is like intersection of two dataframes

df_merge_i = pd.merge(df5, df6, left_on='lKEY', right_on='rKEY',how='inner')
```

```python
df_merge_i
```

```python
# Merging using left,right and outer join
# outer join is like union of two dataframes

df_merge_o = pd.merge(df5, df6, left_on='lKEY', right_on='rKEY',how='outer')
df_merge_o
```

### Defining "agg sum" function

```python
def sum(df3):
  
    sum = df3.agg({
        'Employees': ['sum'],
        'Revenues': ['sum'],
        'Profits': ['sum']
    })
    return sum



sum(df3)
```

### Defining "agg min" function

```python
def min(df3):
  
    min= df3.agg({
        'Employees': ['min'],
        'Revenues': ['min'],
        'Profits': ['min']
    })
    return min



min(df3)
```

```python
def mean(df3):
  
    mean= df3.agg({
        'Employees': ['mean'],
        'Revenues': ['mean'],
        'Profits': ['mean']
    })
    return mean


mean(df3)
```
