### 1.Create a Pandas Series from a Python list, numpy array, and a dictionary.


```python
import numpy as np
import pandas as pd 

list1=[1,2,3,4]
list2=[5,6,7,8]
array1=np.array(list2)
dictionary1={
             'Name':['Malaika','Saman','Sidrah','Tayebah'],
             'University':["NUST","LUMS","GIKI","FAST"],
             'Home town':["Quetta","Peshawar","Karachi","Lahore"]
             }

print("Python list :",list1,"\n")
print("Numpy array",array1,"\n")
print("Dictionary",dictionary1,"\n")

pseries1=pd.Series(list1)
pseries2=pd.Series(array1)
pseries3=pd.Series(dictionary1)

print("Panda series from python list\n",pseries1,"\n")
print("Panda series from Numpy Array\n",pseries2,"\n")
print("Panda series from Dictionary\n", pseries3,"\n")
```

### 2. Assign a custom index to the Series.


```python
custom_index=['a','b','c','d']
pseries1.index=custom_index
print("Series with custom indexing :\n",pseries1)
```

### 3. Perform basic arithmetic operations on Series.


```python
list1=[1,2,3,4]
list2=[5,6,7,8]
array1=np.array(list2)
dictionary1={
             'Name':['Malaika','Saman','Sidrah','Tayebah'],
             'University':["NUST","LUMS","GIKI","FAST"],
             'Home town':["Quetta","Peshawar","Karachi","Lahore"]
             }

pseries1=pd.Series(list1)
pseries2=pd.Series(array1)
pseries3=pd.Series(dictionary1)
sum=pseries1+pseries2
print("Sum of the two pandas series is: \n", sum,"\n")

subtract=pseries2-pseries1
print("Subtraction of the two pandas series is: \n",subtract,"\n")

product=pseries1*pseries2
print("Product of the two pandas series is: \n", product,"\n")

divide=pseries2/pseries1
print("Division of the two pandas series is: \n", divide,"\n")
```

### 4.Access elements using index labels and positions.


```python
print(" Series 1\n",pseries1)
custom_index=['a','b','c','d']
print("\n")
pseries1.index=custom_index
print("Series1 wit custom indexing\n:",pseries1)
access=pseries1.loc['c']
print("\n")
print("Accessing third element of pandas series #01 using index label:",access)
access1=pseries1.iloc[2]
print("Accessing third element of pandas series #01 using position :",access1)
```

### 5.Filter the Series to include only values greater than a specific threshold.


```python
print(pseries1)
filtered_series = pseries1[pseries1 > 2]
print("\n")
print("Series with values greater than 2\n",filtered_series)
```
