 # <span style='color:Red'>  ***What is to Logistic Regression*** </span>


Logistic Regression is a supervised learning algorithm that is used when the target variable is categorical.It accomplishes binary classification tasks by predicting the probability of an outcome, event, or observation. The model delivers a binary or dichotomous outcome limited to two possible outcomes: yes/no, 0/1, or true/false.Logical regression analyzes the relationship between one or more independent variables and classifies data into discrete classes. It is extensively used in predictive modeling, where the model estimates the mathematical probability of whether an instance belongs to a specific category or not.

For example, 0 – represents a negative class; 1 – represents a positive class. Logistic regression is commonly used in binary classification problems where the outcome variable reveals either of the two categories (0 and 1


).

### ***Types of Logistic Regression***

1. **Binary Logistic Regression**:
The dependent variable has only two 2 possible outcomes/classes.
Example-Male or Female.

2. **Multinomial Logistic Regression**:
The dependent variable has only two 3 or more possible outcomes/classes without ordering.
Example: Predicting food quality.(Good,Great and Bad).

3. **Ordinal Logistic Regression**:
The dependent variable has only two 3 or more possible outcomes/classes with ordering. Example: Star rating from 1 to 5

### ***Examples Of Logistic Regression***

Some examples of such classifications and instances where the binary response is expected or implied are:

1. Determine the probability of heart attacks: With the help of a logistic model, medical practitioners can determine the relationship between variables such as the weight, exercise, etc., of an individual and use it to predict whether the person will suffer from a heart attack or any other medical complication.

2. Possibility of enrolling into a university: Application aggregators can determine the probability of a student getting accepted to a particular university or a degree course in a college by studying the relationship between the estimator variables, such as GRE, GMAT, or TOEFL scores.

3. Identifying spam emails: Email inboxes are filtered to determine if the email communication is promotional/spam by understanding the predictor variables and applying a logistic regression algorithm to check its authenticity.

### ***Why is it called regression if it is a classification algorithm?***
The term “regression” in logistic regression refers to the fact that the model is trying to predict the probability of a categorical outcome, such as whether or not someone will buy a product. This is in contrast to linear regression, which is used to predict a continuous outcome, such as the price of a product.

In logistic regression, the probability of the outcome is represented by a logistic function. The logistic function is a sigmoid curve that has a range of values from 0 to 1. This means that the predicted probability of the outcome can be any value between 0 and 1.

# 1. ***Perceptron Trick***

### ***Importing Necessary Libraries***

```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
import numpy as np
```

### ***Generating random  classification data set***


```python
# n_samples means we have data set of 100 numbers 
# n_features (we have 2 input columns)
# n_Informative features are  useful for predicting the target variable. In this case, 1 out of the 2 features will be informative.
# n_redundant features are generated as linear combinations of the informative features. Here, there will be 0 redundant features.
# n_classes number of classes (distinct target labels). Here, the dataset will have 2 classes (binary classification).
# n_Clusters are groups of data points within each class that are close to each other. Here, each class will have 1 cluster.
# n_random_state Sets the seed for the random number generator, ensuring that the same data is generated each time the code is run. This makes the results reproducible.
# hypercube=False Specifies whether to distribute the features uniformly in a hypercube (when True) or within a random Gaussian-like distribution (when False). Setting this to False means the features will not be uniformly distributed.
# class_sep Specifies the factor multiplying the hypercube size. Larger values spread out the classes more and make classification easier. Here, a large class separation value of 10 means the classes will be very well-separated, making the classification task easier.

X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=20)
```

### ***Plotting the classification data***

```python
# green shows those students who will get placement 
# blue shows those students who wont get placement 
# input has 2 features 
# output colmn is represent by colors 

plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

### ***Making perceptron function*** 
 From given input data perceptron function returns weights (coefficients) and intercept


```python

def perceptron(X,y):
    
    # insert a (bias term) new column of ones at the beginning (the first column) of the matrix X  
    # o means index, we are adding 1 on first index, axis=1 along column
    X = np.insert(X,0,1,axis=1)
    
    # By passing X.shape[1] to np.ones, we are creating an array with the same number of elements as there are columns (features) in X.
    # the ouput is intercept w0 , w1 and w2 
    weights = np.ones(X.shape[1])

    # step sixe, learning rate 
    lr = 0.1

    # running loop with epoch value: 1000(times)
    for i in range(1000):
        # selecting a random value between 0 to 100,each time loop runs since shape of x is 100.
        j = np.random.randint(0,100)
        # predicting whether this gives positive or negative number 
        y_hat = step(np.dot(X[j],weights))
        # Updating the weight 
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
        
```

### ***Creating Step Function***

```python
#  step function is created.It reuterns 1 if z is positive and 0 if z is negative.

def step(z):
    if z>0:
       return 1 
    else:
       return 0
```

### ***Determining Value of Coeffients/Weights and Intercepts***

```python
# Calling the function to give  value of coefficient and intercept
intercept_,coef_ = perceptron(X,y)
```

### ***Coeffients/Weights***

```python
print("Weights/Coefficients using perceptron trick is [w1, w2] = ",coef_)

# output is w1 and w2
```

### ***Intercepts***

```python
print("Intercept using perceptron trick is",intercept_)

# output is w0
```

### ***Determining Value of m and b***

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
# np.linspace(-3, 3, 100) creates an array of 100 values starting from -3 and ending at 3. The values are evenly spaced across this range.
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python

plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)

plt.title("Logistic Regression Line Using Perceptron Trick",pad=20,fontsize=16)
# Setting the y-axis Limits.This ensures that the plot focuses on the region where the decision boundary and data points are most relevant.
plt.ylim(-3,2)
```

# 3. ***Sigmoid Function***
We need to make some changes in perceptron trick to achieve the same results as scikit learn.Here comes sigmoid function.Logistic regression uses a logistic function called a sigmoid function to map predictions and their probabilities. The sigmoid function refers to an S-shaped curve that converts any real value to a range between 0 and 1.

# Sigmoid Function




Moreover, if the output of the sigmoid function (estimated probability) is greater than a predefined threshold on the graph, the model predicts that the instance belongs to that class. If the estimated probability is less than the predefined threshold, the model predicts that the instance does not belong to the class.

For example, if the output of the sigmoid function is above 0.5, the output is considered as 1. On the other hand, if the output is less than 0.5, the output is classified as 0. Also, if the graph goes further to the negative end, the predicted value of y will be 0 and vice versa. In other words, if the output of the sigmoid function is 0.65, it implies that there are 65% chances of the event occurri.




where:
e: The base of the natural logarithm (approximately 2.71828)
x: The input value
σ(x): The output value between 0 and 1e.

In perceptron trick we used step function.However,here We will use sigmoid function It scaled down numbers between 0 to 1
From step to sigmoid we have converted the system into probablistic interpretation



### Change in perceptron equation after replacing step with sigmoid function 




### ***Making Sigmoid function***

```python
def perceptron(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = sigmoid(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
```

## ***Cost Function/Error Function***

Cost function is a function that measures the performance of a Machine Learning model for given data.
Cost Function is basically the calculation of the error between predicted values and expected values and presents it in the form of a single real numberDifference between cost and loss function is that rms Cost Function is the average of error of n-sample in the data and Loss Function is the error for individual data points.In other words,Loss Function is for one training example,Cost Function is the for the entire training .ad
It can also predicts between two models,which one is bet.tg)



### ***In Cost Function we use two methods:*** ###

1. Maximum Likelihood
2. Cross Entropy

## ***Maximum likelihood***

The logistic regression model is fit to the data using a maximum likelihood estimation (MLE) procedure. The MLE procedure is a statistical method that finds the parameters of a model that maximize the likelihood of the data.
Once the logistic regression model has been fit to the data, it can be used to predict the probability of a binary outcome for new data. The predicted probability is calculated as the output of the model, which is a linear combination of the predictor variables, with the sigmoid function applied to the resu.

To convert the predicted probabilities into actual binary predictions, a threshold is often applied. For example, if the threshold is set at 0.5, any predicted probability above 0.5 is classified as the positive class (y = 1), while probabilities below 0.5 are classified as the negative class (y = 0ates



Concept of maximum likehood says that when we multiply the prob of points and got greater number compared to other line,the line with greater product of probabilities of points is best fit line.


## ***Cross Entropy:***


For large dataset, maximum likelihood is not valid since multiplication of a large number of probabilities will give a small number.
Therefore we take log of this function so it may  convert into sum.This is known as cross entropy (Sum of -ive of log of maximum likeihood)
Cross Entropy to be minimized to give best fit line.We need to find the coefficients of line that gives minimum cross entropy.



To find its solution we have to apply gradient descent

# 4. ***Gradient Descent***

To minimize the Cost Function and find the optimal values of ‘θ’, we use the Gradient Descent algorithm. Gradient Descent iteratively updates the weights based on the gradient of the Cost Function. 



```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=20)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
# No Regularization: penalty='none' shows that it will not penalize large coefficients and 
# will try to fit the training data as closely as possible without any constraint.
# SAG Solver: solver='sag' means the optimization algorithm used will be Stochastic Average Gradient Descent, 
# which is efficient for large datasets by performing gradient descent on a subset of the data in each iteration.

lor = LogisticRegression(penalty='none',solver='sag')
lor.fit(X,y)
```

```python
print(lor.coef_)
print(lor.intercept_)
```

```python
m1 = -(lor.coef_[0][0]/lor.coef_[0][1])
b1 = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m1*x_input + b1
```

### Inside for loop,we have used two equations,they are: 
y_hat is predicted y value by gradient descent alogorithm while weights are coefficients of line.

####         y_hat = sigmoid(np.dot(X,weights))



####         weights = weights + lr*(np.dot((y-y_hat),X)/X.shape[0])



```python
def gd(X,y):

    # Inserting a column of 1 for intercept
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.5
    
    for i in range(10000):
        y_hat = sigmoid(np.dot(X,weights))
        # Using Weight Update rule(gradient descent formula)
        weights = weights + lr*(np.dot((y-y_hat),X)/X.shape[0])
        
   # Returning All Coefficients and Intercept     
    return weights[1:],weights[0]
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
coef_,intercept_ = gd(X,y)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3,label='Logistic Regression using scikit learn')
plt.plot(x_input1,y_input1,color='black',linewidth=3,label='Gradient Descent')
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
plt.title("Logistic Regression using Gradient Descent",pad=15,fontsize=16)

plt.legend()
plt.show()
```

### ***Conclusion***

Using Gradient Descent (shown by Black line), our model has reached almost near to red line (scikit learn method).There's a mimium error left,which can be reduced using greater number of epoch (by running loop more times).Gradient descent symmetrically divides the data points and its by far the best fit line.
