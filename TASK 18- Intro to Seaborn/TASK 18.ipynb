## ðŸŸ¡ Scatter Plot with seaborn


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Life Expectancy Data.csv")
```

```python
df.isnull().sum().sum()
```

```python
df.duplicated().sum()
```

```python
# Replacing infinity values
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
df.head(5)
```

```python
# scatter plot -> axes level function
# Extracting some rows from dataframe for better visualization
df1=df.iloc[850:900]
```

```python
sns.scatterplot(data=df1, x='Country', y='Life expectancy ')
```

```python
sns.scatterplot(data=df1, x='Country', y='Life expectancy ',hue='Status')
```

##  ðŸ”¹ Plotting Developing and Developed countries on different plots


```python
sns.relplot(data=df1, x='Country', y='Life expectancy ',kind='scatter',col='Status')
```

## ðŸ”¹ Adding plot title and x,y labels

```python
print(df.columns)
```

```python
sns.scatterplot(data=df1, x='Country', y='Life expectancy ',hue='Status',style='Status')

plt.title('Life Expectancy by Country and Status (Developing/Developed)',fontsize=14,pad=20)
plt.xlabel('Country', labelpad=20,fontsize=14)
plt.ylabel('Life Expectancy', labelpad=20,fontsize=14)
```

##  ðŸ”¹ Changing Palette


```python
sns.scatterplot(data=df1, x='Country', y='Life expectancy ',hue='Status')

plt.title('Life Expectancy by Country and Status (Developing/Developed)',fontsize=14,pad=20)
plt.xlabel('Country', labelpad=20,fontsize=14)
plt.ylabel('Life Expectancy', labelpad=20,fontsize=14)

sns.set_palette('Set1') 
```

```python
sns.scatterplot(data=df1, x='Country', y='Life expectancy ',hue='Status')

plt.title('Life Expectancy by Country and Status (Developing/Developed)',fontsize=14,pad=20)
plt.xlabel('Country', labelpad=20,fontsize=14)
plt.ylabel('Life Expectancy', labelpad=20,fontsize=14)

sns.set_palette('Paired')
```

## ðŸŸ¡ Line Plot with seaborn


```python
sns.lineplot(data=df, x='Year', y='Life expectancy ', errorbar=None)
plt.title('Change of Life Expectancy \n in countries over the years \n 2000-2014' ,fontsize=14,pad=20)

plt.xlabel('Year', labelpad=20,fontsize=14)
plt.ylabel('Life Expectancy', labelpad=20,fontsize=14)

# errobar=None shows that i want output without the light shaded area.
```

```python
# EXTRAXTING DATA OF 3 COUNTRIES TO SEE THE LIFE EXPECTENCY OVER THE YEARS AND PLOT THEM IN A SINGLE PLOT 

df2=df[df['Country'].isin(['Pakistan','Austria','India','Ireland'])]
df[['Country','Status']].value_counts()
```

```python
sns.lineplot(data=df2, x='Year', y='Life expectancy ',hue='Country',errorbar=None,style='Status')
plt.title('Change of Life Expectancy \n in 4 countries over the years \n 2000-2014' ,fontsize=14,pad=20)

plt.xlabel('Year', labelpad=20,fontsize=14)
plt.ylabel('Life Expectancy', labelpad=20,fontsize=14)
```

```python
df3=df[df['Country'].isin(['Pakistan','Austria','India'])]
```

```python
sns.relplot(data=df3, x='Year', y='Life expectancy ',kind='line',col='Country',style='Status')
plt.subplots_adjust(top=0.9)
sns.set_palette('Paired')

```

## ðŸŸ¡ Bar Plot with seaborn


```python
sns.barplot(data=df1, x='Country', y='Adult Mortality',errorbar=None)

plt.title('Adult Morality Rate in 5 Countries over 14 years (2000-2014)' ,fontsize=14,pad=20)

plt.xlabel('Country', labelpad=20,fontsize=14)
plt.ylabel('Adult Mortality', labelpad=20,fontsize=14)

```

```python
sns.barplot(data=df1, x='Country', y='Adult Mortality',hue='Status',errorbar=None)

plt.title('Adult Morality Rate in 5 Countries over 14 years (2000-2014)' ,fontsize=14,pad=20)

plt.xlabel('Country', labelpad=20,fontsize=14)
plt.ylabel('Adult Mortality', labelpad=20,fontsize=14)

```

## ðŸŸ¡ Count Plot with seaborn


```python
dfy=df[df['Year'].isin([2014])]
dfy
```

```python
sns.countplot(x='Status',data=dfy)
plt.title('Developed and Developing countries around the world in year 2014' ,fontsize=14,pad=20)

plt.xlabel('Country', labelpad=20,fontsize=14)
plt.ylabel('Count', labelpad=20,fontsize=14)
```

```python

sns.countplot(x='Status',data=df1,hue='Country')
plt.title('Ratio of 5 random countries categorized as \n Developed and Developing over 14 years span' ,fontsize=14,pad=20)

plt.xlabel('Country', labelpad=20,fontsize=14)
plt.ylabel('Count', labelpad=20,fontsize=14)
```

## ðŸŸ¡ Box Plot with seaborn
A boxplot is a standardized way of displaying the distribution of data based on a five number summary (â€œminimumâ€, first quartile [Q1], median, third quartile [Q3] and â€œmaximumâ€). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed

```python
sns.boxplot(data=df2,x='Country',y='Life expectancy ')
```

```python
sns.boxplot(data=df2,x='Country',y='Life expectancy ',hue='Status')
plt.title('Comparison of Average Life Expectency in \n  developed and  developing countries \n (Acc to Statistics 2000-2014)',fontsize=14,pad=20)

plt.xlabel('Country', labelpad=20,fontsize=14)
plt.ylabel('Life Expectency', labelpad=20,fontsize=14)
```
