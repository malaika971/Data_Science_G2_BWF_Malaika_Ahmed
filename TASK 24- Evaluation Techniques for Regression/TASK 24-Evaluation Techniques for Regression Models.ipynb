Evaluation techniques for regression models are crucial for assessing their performance. Common metrics include Mean Absolute Error (MAE), which measures the average magnitude of errors, and Mean Squared Error (MSE), which emphasizes larger errors due to squaring the differences. Root Mean Squared Error (RMSE) provides error estimates in the same unit as the target variable. R-squared (RÂ²) indicates the proportion of variance explained by the model, while Adjusted R-squared accounts for the number of predictors. Mean Absolute Percentage Error (MAPE) offers a percentage-based accuracy metric, and Median Absolute Error provides a robust measure of central tendency for errors, less influenced by outliers. These metrics help in comparing and refining regression models.

 # <span style='color:Red'> ***First, creating Linear Regression Model*** </span>



```python
# Importing Libraries
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.metrics import mean_absolute_percentage_error, median_absolute_error
```

### ***Importing Data Set***

```python
df = pd.read_csv('pckg.csv')
```

```python
df.head()
```

```python
df.shape
```

### ***Creating Scatter plot between cgpa and package***

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

### ***Separating Input and Target columns***

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

### ***Splitting Data into training and testing***


```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

### ***Defining an object for linearRegression***


```python
lr = LinearRegression()
```

### ***Fitting/Training the Linear Regression model***


```python
lr.fit(X_train,y_train)
```

```python
plt.scatter(df['cgpa'],df['package'],color='g')

plt.plot(X_train,lr.predict(X_train),color='red')

plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

- It is Average of the squared differences between predicted and actual values.

- The MSE is a measure of the quality of an estimator â€” it is always non-negative, and values closer to zero are better.
- Sensitive to outliers, as squaring emphasizes large errors.

 ##  <span style='color:Blue'> ***4. R2 Score*** </span>

- R2 score is also called "Coefficient of Determinant" and "Goodness of Fit"
- The R2 score (pronounced R-Squared Score) is a statistical measure that tells us how well our model is making all its predictions on a scale of zero to one.
- We can use the R2 score to determine the accuracy of our model in terms of distance or residual.
- Measures the proportion of variance in the dependent variable explained by the model.
- Values closer to 1.0 indicate a better model.





 





### ***Advantage***
- Easy to interpret results.If R2 score is 1,suggests that Regression line perfectly fits the model,making no error.If R2 is zero,it suggests that Regression line is same as mean line.Input column (cgpa) is making no impact in predicting salary.Model is predicting the salary jusy by calculating average salary.Indicates poor model.

### ***Limitation***
- Can increase artificially as you add more features, even if unnecessary.To deal with it we have adjusted R2 score




## ***Calculating R2 SCORE***


```python
print("R2 SCORE",r2_score(y_test,y_pred))
```

 ##  <span style='color:Blue'> ***5. Adjusted R2 Score*** </span>
- The adjusted R-squared is a modified version of R-squared that adjusts for predictors that are not significant in a regression model.- 
Compared to a model with additional input variables, a lower adjusted R-squared indicates that the additional input variables are not adding value to the model
- 
Compared to a model with additional input variables, a higher adjusted R-squared indicates that the additional input variables are adding value to the mode




### ***When to use Adjusted R2 Score***
When we are having more input columns/Multiple linear regression ,adjusted R2 score is good to use.It helps in predicting whether input column is relevant or irrevant. l.

## ***Calculating Adjusted R2 SCORE***


```python
X_test.shape
```

```python
r2 = r2_score(y_test,y_pred)
```

```python
# n is number of observations thats 40 in our case.
# k is input columns ,in our case its 1 (cgpa)
#1 - ((1-r2)*(n-1)/(n-k-2))

ad=1 - ((1-r2)*(40-1)/(40-1-1))
print("Adjusted R2 Square value is :",ad)
```

 ##  <span style='color:Blue'> ***6. Mean Absolute Percentage Error (MAPE)*** </span>
- Mean absolute percentage error (MAPE) is the percentage equivalent of mean absolute error (MAE).
- Mean absolute percentage error measures the average magnitude of error produced by a model, or how far off predictions are on average.
- A MAPE value of 20% means that the average absolute percentage difference between the predictions and the actuals is 20%. In other words, the modelâ€™s predictions are, on average, off by 20% from the real values.
- A lower MAPE value indicates a more accurate prediction â€“ an MAPE of 0% means the prediction is the same as the actual.
- While a higher MAPE value indicates a less accurate prediction.
