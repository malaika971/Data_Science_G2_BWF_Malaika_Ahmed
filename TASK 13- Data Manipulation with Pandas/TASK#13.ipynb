## 8.Group the DataFrame by a specific column and calculate the mean of each group.


```python
groupby_data=company.groupby('Sector')
print(groupby_data)

mean=groupby_data.mean(numeric_only=True) 
print(mean)
```

## 9. Group the DataFrame by multiple columns and calculate the sum of each group.


```python
groupby_data=company.groupby(['Sector','Industry'])
print(groupby_data)

```

```python
sum=groupby_data.sum(numeric_only=True) 
print(sum)
```

## 10. Use the agg method to apply multiple aggregation functions to grouped data.


```python
groupby_data = company.groupby('Sector')
agg_data = groupby_data.agg({
    'Employees': ['mean', 'sum', 'max', 'min'],
    'Revenues': ['mean', 'sum', 'max', 'min']
})

print(agg_data)
```

## 11. Calculate the size of each group.


```python
groupby_data=company.groupby('Sector')
print(groupby_data)
Sector=groupby_data.size()
print(Sector)
```
