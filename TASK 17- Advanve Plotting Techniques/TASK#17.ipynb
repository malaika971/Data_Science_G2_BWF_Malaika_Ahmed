## 2. Customize the subplots with titles, axis labels, and legends.

```python
df1=df['Temperature_C']
df2=df['Humidity_pct']
df3=df['Precipitation_mm']
df4=df['Wind_Speed_kmh']
df5=df['Location']

x=df1[0:5]
x1=df5[0:5]
y1=df2[0:5]
y2=df3[0:5]
y3=df4[0:5]
y4=df1[0:5]
```

```python

# 4 SUBLOTS

plt.suptitle("Multiple Plots with Titles and Labels ",fontsize=16)

# LINE PLOT
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.xlabel("Temperature (C)")
plt.ylabel("Humidity (pct)")
plt.title("Line Plot")


# SCATTER PLOT
plt.subplot(2,2,2)
color=['red','blue','green','brown','orange']
size = [180, 150, 250, 125, 800]
plt.scatter(x, y2, marker='.', c=color, s=size)
plt.xlabel("Temperature (C)")
plt.title("Scatter plot")
plt.ylabel("Precipitation (mm)")


# HORIZONTAL BAR CHART
plt.subplot(2,2,3)
plt.barh(x1,y3)
plt.xlabel("Location")
plt.ylabel("Wind_Speed (kmh)")
plt.title("Bar Plot")


# PIE CHART
plt.subplot(2,2,4)
plt.pie(y3,labels=x1,autopct='%2.1f%%')
plt.title("Pie Chart")
plt.subplots_adjust(left=0.1, right=0.9, top=0.8 , bottom=0.1, wspace=0.5, hspace=0.9)


plt.show()
```

## ðŸŸ¡ HISTOGRAMS


### 1. Create a histogram for a given numerical data set. Customize the number of bins.


```python
Marks = [12, 15, 22, 35, 37, 45, 50, 55, 65, 68, 70, 75, 80, 85, 90, 95, 100, 42, 58, 78]

plt.hist(Marks,5, histtype='bar', rwidth=0.8)
plt.xlabel("Marks Range")
plt.ylabel("No of Students")
plt.title("Histogram for single numerical data set \n with gap in bars",fontsize=14,pad=25)

plt.show()
```

## ðŸ”¹ Histogram with Different Edge Color

```python
Marks = [12, 15, 22, 35, 37, 45, 50, 55, 65, 68, 70, 75, 80, 85, 90, 95, 100, 42, 58, 78]

plt.hist(Marks,3, histtype='bar', ec='r')
plt.xlabel("Marks Range")
plt.ylabel("No of Students")
plt.title("Histogram with edge Color: Red",fontsize=14,pad=20)
plt.show()
```

## ðŸ”¹ Histogram with colored bars

```python
import matplotlib.pyplot as plt

Marks = [12, 15, 22, 35, 37, 45, 50, 55, 65, 68, 70, 75, 80, 85, 90, 95, 100, 42, 58, 78]

plt.hist(Marks, bins=[1,11,21,31,41,51,61,71,81,91,101] , histtype='bar', ec= 'g', fc= 'y')
plt.xlabel("Marks Range")
plt.ylabel("No of Students")
plt.title("Histogram with colored bars",fontsize=14,pad=20)
plt.show()
```

# ðŸŸ¡ Box Plots


### 1. Create a box plot for a given numerical data set


```python
df1=df['Temperature_C']


plt.boxplot(df1)

plt.ylabel("Temperature in Celsius")
plt.title("Box Plot")

plt.show()

```

### ðŸ”¹ Horizontal Box Plot

```python
df1=df['Temperature_C']


plt.boxplot(df1,vert=False)

plt.xlabel("Temperature in Celsius")
plt.title("Horizontal Box Plot" )

plt.show()

```

### ðŸ”¹ Horizontal Box with varying width

```python
# changing width - default width is 0.2

df1=df['Temperature_C']


plt.boxplot(df1,widths=0.5)

plt.ylabel("Temperature in Celsius")
plt.title("Box Plot of the Data Set")

plt.show()

```

### ðŸ”¹ Horizontal Box with filled color

```python
# FILLING COLOR

df1=df['Temperature_C']


plt.boxplot(df1,patch_artist=True)

plt.ylabel("Temperature in Celsius")
plt.title("Box Plot of the Data Set")

# Show the plot
plt.show()

```

### ðŸ”¹ Horizontal Box with visible mean

```python
# Showing the mean of data
df1=df['Temperature_C']


plt.boxplot(df1,showmeans=True)

plt.ylabel("Temperature in Celsius")
plt.title("Box Plot showing mean in green triangle")

plt.show()

```

### ðŸ”¹ Horizontal Box With Visible Outlier

```python


plt.boxplot(df1)

plt.ylabel("Temperature in Celsius")
plt.title("Box Plot with outlier")

plt.show()

```

### ðŸ”¹ Box Plot Showing whisker meeting the Outlier 

```python

df1=df['Temperature_C']


plt.boxplot(df1,whis=4)

plt.ylabel("Temperature in Celsius")
plt.title("Box Plot of the Data Set")

# Show the plot
plt.show()

```

###  ðŸ”¹ Box Plot With Varying Outlier (Color and Symbol) 

```python
# changing color of outlier and changing its shape
df1=df['Temperature_C']


plt.boxplot(df1,sym='g*')

plt.ylabel("Temperature in Celsius")
plt.title("Box Plot with green color outlier and * symbol")

plt.show()

```

###  ðŸ”¹ Box Plot With Varying Caps Color

```python
 # ADDING COLOURS

df2=df['Humidity_pct']
plt.boxplot(df2,capprops=dict(color='g'))

plt.ylabel("Humidity (pct)")
plt.title("Box Plot Showing Green Colored Caps",fontsize=14,pad=20)

plt.show()

```

###  ðŸ”¹ Box Plot With Varying Whiskers Color

```python
 # ADDING COLOURS
df2=df['Humidity_pct']


plt.boxplot(df2,whiskerprops=dict(color='b'))

plt.ylabel("Humidity (pct)")
plt.title("Box Plot Showing blue Whiskers",fontsize=14,pad=20)

plt.show()

```

###  2.Generate multiple box plots in a single figure to compare different data sets.


```python

# Generate multiple box plots in a single figure to compare different data sets.

df1=df['Temperature_C']
df2=df['Humidity_pct']
df3=df['Precipitation_mm']
df4=df['Wind_Speed_kmh']

plt.boxplot([df1, df2,df3,df4])
plt.show()
```

## 3.Customize the box plots with titles, labels, and change the colors.


```python

df1=df['Temperature_C']
df2=df['Humidity_pct']
df3=df['Precipitation_mm']
df4=df['Wind_Speed_kmh']


plt.boxplot([df1, df2,df3,df4], labels=['Temperature ', 'Humidity', 'Precipitation', 'Wind Speed'])

plt.title("Multiple Box Plots",fontsize=14,pad=20)
plt.show()
```

```python

df1=df['Temperature_C']
df2=df['Humidity_pct']
df3=df['Precipitation_mm']
df4=df['Wind_Speed_kmh']

whiskerprops = dict(color="blue", linewidth=1.5)
capprops = dict(color="red", linewidth=1.5)

plt.boxplot([df1, df2,df3,df4], labels=['Temperature ', 'Humidity', 'Precipitation', 'Wind Speed'],whiskerprops=whiskerprops,
            capprops=capprops)

plt.title("Multiple Box Plots With Varying Color Whiskers and Caps",fontsize=14,pad=20)
plt.show()
```
