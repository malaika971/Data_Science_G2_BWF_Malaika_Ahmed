***What is Regression?***

Regression is a statistical technique used to analyze and model the relationship between a dependent variable (also called the response or outcome variable) and one or more independent variables (also called predictors or explanatory variables). In simple terms, regression allows you to find the best-fit line or curve that describes the relationship between variablesThe dependent variable is the one that is being predicted or modeled, while the independent variables are used to predict the dependent variable.

***Why Regression?***

- The purpose of regression analysis is to determine the strength of the relationship between variables and to make predictions about the dependent variable based on the values of the independent variables
- Regression can be used for prediction, forecasting, and identifying the factors that influence the outcome variable.

  .


### ***Linear Regression:***



- Linear regression is used to model the relationship between a dependent variable and one or more independent variables, where the relationship is assumed to be linear.
-  The goal of linear regression is to find the linear equation that best describes the relationship between the variables.
-  Linear Regression finds the line of best fit that minimizes the sum of the squared distances between the observed values of the dependent variable and the predicted values by the model.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('pckg.csv')
```

```python
df.head()
```

```python
# plotting cgpa and packge to see thier linear relationship

plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
# Separating x input and y output columns
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
# Splitting data into test and train 

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# fit is used for training the data 

lr.fit(X_train,y_train)
```

***

### **Plotting the train data**

```python

plt.scatter(df['cgpa'],df['package'],color='b',label="Data points")
plt.plot(X_train,lr.predict(X_train),color='red',label="Regression Line")

plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```
